```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# order the 'Stay' column.
df["Stay"] = pd.Categorical(df["Stay"], 
                      categories=["0-10","20-Nov","21-30","31-40", "41-50", "51-60", "61-70", "71-80", "81-90", "91-100"],
                      ordered=True)
```

```python
# order the 'Type of Admission' column.
df["Type of Admission"] = pd.Categorical(df["Type of Admission"], 
                      categories=["Trauma","Urgent","Emergency"],
                      ordered=True)

df["Type of Admission"]=df["Type of Admission"].cat.codes
```

```python
# order the 'Severity of Illness' column.
df["Severity of Illness"] = pd.Categorical(df["Severity of Illness"], 
                      categories=["Minor","Moderate","Extreme"],
                      ordered=True)
df["Severity of Illness"]=df["Severity of Illness"].cat.codes
```

```python
# order the 'Department' column.
df["Department"] = pd.Categorical(df["Department"], 
                      categories=["gynecology","radiotherapy","anesthesia","TB & Chest disease"],
                      ordered=True)

df["Department"]=df["Department"].cat.codes
```

```python
# order the 'Age' column.
df["Age"] = pd.Categorical(df["Age"], 
                      categories=["0-10","11-20","21-30","31-40","41-50","51-60","61-70","71-80","81-90","91-100"],
                      ordered=True)

df["Age"]=df["Age"].cat.codes
```

```python
# Plot the "Type of Admission"
sns.pointplot(x="Stay", y="Type of Admission", data=df)
plt.legend(labels=['0=Trauma', '1=Urgent', '2=Emergency'])
```
