```python

df["Stay"] = pd.Categorical(df["Stay"], 
                      categories=["0-10","20-Nov","21-30","31-40", "41-50", "51-60", "61-70", "71-80", "81-90", "91-100", "More than 100 Days"],
                      ordered=True)
df["Stay"]=df["Stay"].cat.codes
```

```python
df['Stay'] = df['Stay'].replace({0 : '0-10', 1 : '11-20', 2 : '21-30', 3 : '31-40', 4 : '41-50', 5 : '51-60', 6 : '61-70', 7 : '71-80', 8 : '81-90', 9 : '91-100', 10 : 'More than 100'})
```

```python
df['Stay']=sorted(df['Stay'])
```

```python
df['Stay']
```

```python
df["Stay_Labels"] = pd.Categorical(df["Stay"], 
                      categories=["0-10","11-20","21-30","31-40", "41-50", "51-60", "61-70", "71-80", "81-90", "91-100", "More than 100"],
                      ordered=True)
df["Stay_Labels"]=df["Stay_Labels"].cat.codes
```

```python

df["Severity of Illness"] = pd.Categorical(df["Severity of Illness"], 
                      categories=["Minor","Moderate","Extreme"],
                      ordered=True)
df["Severity of Illness"]=df["Severity of Illness"].cat.codes
```

```python
df['Severity of Illness'] = df['Severity of Illness'].replace({0 : '1-Minor', 1 : '2-Moderate', 2 : '3-Extreme'})
```

```python
df['Severity of Illness']=sorted(df['Severity of Illness'])
```

```python
df["Severity of Illness Labels"] = pd.Categorical(df["Severity of Illness"], 
                      categories=["1-Minor","2-Moderate","3-Extreme"],
                      ordered=True)
df["Severity of Illness Labels"]=df["Severity of Illness Labels"].cat.codes
```

```python
from scipy.stats import chi2_contingency
```

```python
stat, p, dof, expected = chi2_contingency(df['Available Extra Rooms in Hospital'], ['City_Code_Hospital']) 
alpha = 0.05
print("p value is " + str(p)) 
if p <= alpha: 
    print('Dependent (reject H0)') 
else: 
    print('Independent (H0 holds true)') 
```

```python

# defining the table 
data = [[207, 282, 241], [207, 282, 241], [322,335,655]] 
stat, p, dof, expected = chi2_contingency(data) 
  
# interpret p-value 
alpha = 0.05
print("p value is " + str(p)) 
if p <= alpha: 
    print('Dependent (reject H0)') 
else: 
    print('Independent (H0 holds true)')
```
