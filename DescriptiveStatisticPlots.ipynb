```python
dft1=pd.crosstab(df['Department'], df['Stay'])
dft1
```

```python
chiStats = stats.chi2_contingency(observed= dft1)
```

```python
print(chiStats[1], chiStats[2])
```

```python
alpha = 0.05
critical_value = crit = stats.chi2.ppf(q = 1 - alpha,
                      df = chiStats[2])
```

```python
observed_chi_val = chiStats[0]
```

```python
print(observed_chi_val, critical_value)
```

```python
if observed_chi_val <= critical_value:
    print ('Null hypothesis cannot be rejected (variables are independent)')
else:
    print ('Null hypothesis can be rejected (variables are dependent)')
```

```python
import numpy as np
from scipy import stats
N = 10000
a = df_label['Bed Grade']
b = df_label['Stay']
var_a = a.var(ddof=1)
var_b = b.var(ddof=1)
s = np.sqrt((var_a + var_b)/2.0)
t = (a.mean() - b.mean())/(s*np.sqrt(2.0/N))
df = len(a) + len(b) - 2.0
p = stats.t.cdf(t,df=df)
print("t = " + str(t))
print("p = " + str(2.0*p))
```
