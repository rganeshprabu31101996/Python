```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\ghpu\JupyterNotebook\datasets\healthcare\healthdata.csv")
```

```python
df
```

```python
import scipy.stats as stats
```

```python
from sklearn.preprocessing import LabelEncoder
df=df.apply(LabelEncoder().fit_transform)
```

```python
dft1=pd.crosstab(df['case_id'], df['Stay'])
dft1
```

```python
chiStats = stats.chi2_contingency(observed= dft1)
```

```python
print(chiStats[1], chiStats[2])
```

```python
alpha = 0.05
critical_value = crit = stats.chi2.ppf(q = 1 - alpha,
                      df = chiStats[2])
```

```python
observed_chi_val = chiStats[0]
```

```python
print(observed_chi_val, critical_value)
```

```python
if observed_chi_val <= critical_value:
    print ('Null hypothesis cannot be rejected (variables are independent)')
else:
    print ('Null hypothesis can be rejected (variables are dependent)')
```
